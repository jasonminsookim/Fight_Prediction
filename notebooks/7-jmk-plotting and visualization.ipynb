# 1. Import Libraries and Dataset

```python
import pandas as pd
import seaborn as sns; sns.set()
import re
import matplotlib.pyplot as plt

# Import Data
df = pd.read_csv("../data/scores/hopt_scores_v3.csv")
```

# 2. Wrangle Data

```python
for index, row in df.iterrows():
    df.loc[index, 'lookback_window'] = int(re.sub("[^0-9]", "", row['dataset']))
    df.loc[index, 'aggr_type'] = row['dataset'].split(":")[0]
```

# 3.1 Plot Cumulative Data Non Feature Engineered

### Non-FE Cumulative

```python
# subset data so that it only shows for 
cumulative_df = df.loc[df['aggr_type'] == "Cumulative Data"]
cumulative_df = cumulative_df[cumulative_df['dict_type'] == "cumu_dfs_dict"]

a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

cumulative_plot = sns.lineplot(x="lookback_window", y="accuracy", hue="model_name", data = cumulative_df, ax = ax)
cumulative_plot = cumulative_plot.set_title("Accuracy of Various Machine Learning Models using non-FE Cumulative Data")
```

```python
fig.savefig('../data/plots/hyperopt/nonFE_cumulative_v1.png') # Saves the above plot
```
